```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tracks = pd.read_csv('/content/drive/MyDrive/datasets/tracks.csv')
```

```python
# top 10 least popular songs
least_popular_tracks = tracks.sort_values('popularity', ascending=True).head(10)
```

```python
# top 10 popular songs using query and sort
most_popular_tracks = tracks.query('popularity > 90', inplace=False).sort_values('popularity', ascending=False)
```

```python
most_popular_tracks[:10]
```

```python
# changing the index inplace
tracks.set_index('release_date', inplace=True)
tracks.index = pd.to_datetime(tracks.index)
```

```python
tracks.head()
```

```python
tracks.info()
```

```python
tracks[['artists']].iloc[18]
```

```python
# chaning duration ms to s
tracks['duration'] = tracks['duration_ms'].apply(lambda x: round(x/1000))
tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
tracks['duration'].iloc[0]
```

```python
tracks['duration'].head()
```

```python
correlation_df = tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(correlation_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='magma', linewidths=1, linecolor='Black')
heatmap.set_title('Correlation heatmap of features')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_tracks = tracks.sample(int(0.004 * len(tracks)))
```

```python
sample_tracks
```

```python
sample_tracks.shape
```

```python
plt.figure(figsize=(10, 6))
# regression plot
sns.regplot(data=sample_tracks, y='loudness', x = 'energy', color='r').set_title('Loudness vs Energy correlation')
```

```python
plt.figure(figsize=(10, 6))
# regression plot
sns.regplot(data=sample_tracks, y='popularity', x = 'acousticness', color='r').set_title('Popularity vs Acousticness correlation')
```

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=years, binwidth=3, discrete=True).set(title="Number of songs released per year")
```

```python
figure_dimensions = (18, 7)
fig, ax = plt.subplots(figsize = figure_dimensions)
fig = sns.barplot(x = years, y = tracks['duration'], ax=ax, errwidth=False).set(title = "Year vs Duration")
plt.xticks(rotation=90)
```

```python
tracks_durations = tracks['duration']
sns.set_style(style='whitegrid')
fig, ax = plt.subplots(figsize=(10, 5))
fig = sns.lineplot(x = years, y = tracks_durations, ax = ax).set(title = "Years vs Durations")
plt.xticks(rotation=30)
```
